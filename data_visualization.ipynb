# Data Visualization

```python
#import all necessary packages
import pandas as pd
import numpy as np
import pandas_profiling as pp
import matplotlib.pyplot as plt 
import seaborn as sb
```

```python
#load dataset
data=pd.read_csv("Dataset/train-data.csv",encoding='ANSI')
```

```python
data.head()
data['Power']=data['Power(bhp)']
```

```python
#drop unusualcolumn
data.drop('S_number',axis=1,inplace=True)
data.head()
```

```python
#shape of data before removing duplicates
data.shape
```

```python
#Removing duplicate element
data_dup=data.drop_duplicates()
data_dup.head()
```

```python
data_dup['Price']=data_dup['Price']*100000
data_dup.head()
```

```python
#remove blank rows
data_dup.dropna(axis=0,how='all',inplace=True)
data_dup.isnull().sum()
```

```python
data_dup.shape
data_dup.drop(["New_Price","Power(bhp)"],axis=1,inplace=True)
data_dup.fillna(0,inplace=True)
```

```python
data1=data_dup
data1.head()
```

# Handle categorical variables with numerically coded values


```python
data_dup.dtypes
```

```python
#converting columns to appropriate data type
data_dup['Transmission']=data_dup.Transmission.astype('category')
data_dup['Engine(CC)']=data_dup['Engine(CC)'].astype('int64')
data_dup['Seats']=data_dup['Seats'].astype('int64')
```

```python
data_dup.dtypes
```

```python
#Create numeric based column for variable which are of object dtype by creating ID's for each category object
data_dup['Name_id']=pd.Categorical(data_dup.Name).codes
data_dup['Loction_id']=pd.Categorical(data_dup.Location).codes
data_dup['Fuel_Type_id']=pd.Categorical(data_dup.Fuel_Type).codes
data_dup['Transmission_id']=pd.Categorical(data_dup.Transmission).codes
data_dup['Owner_Type_id']=pd.Categorical(data_dup.Owner_Type).codes
data_dup['Power_id']=pd.Categorical(data_dup.Power).codes
data_dup['type_id']=pd.Categorical(data_dup.type).codes
data_dup['condition_id']=pd.Categorical(data_dup.condition).codes
data_dup.head()
```

# Perform univariate and bivariate analysis and derive meaningful insights about the dataset

```python
#Univariate Analysis that are perform on categorical data and numerical data by plotting histogram
```

```python
#Univariate Analysis on Categorical data
data_dup['Transmission'].value_counts().plot.bar(title='Transmission graph')
plt.grid(color='r')
plt.show()
```

```python
#Univariate Analysis for numerical data
col_names = ['Year','Kilometers_Driven','Mileage(kmpl)','Engine(CC)','Seats','Price']

fig, ax = plt.subplots(len(col_names), figsize=(16,12))

for i, col_val in enumerate(col_names):

    sb.distplot(data_dup[col_val], hist=True, ax=ax[i])
    ax[i].set_title('Freq dist '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=16)
    ax[i].set_ylabel('Count', fontsize=8)
plt.show()
```

```python
#Bivariate Analysis used to find the relationship between two variable by analyzing the scatter plot
car_data_hist = data1.drop(['Name','Location','Fuel_Type','Transmission','Owner_Type','paint_color','type','condition','Power'],axis=1)
sb.pairplot(car_data_hist)
plt.show()
```

# Removing Outliers

```python
#outlier detection analysis
col_names =  ['Year','Kilometers_Driven','Mileage(kmpl)','Engine(CC)','Seats','Price']

fig, ax = plt.subplots(len(col_names), figsize=(8,40))

for i, col_val in enumerate(col_names):

    sb.boxplot(y=car_data_hist[col_val], ax=ax[i])
    ax[i].set_title('Box plot - {}'.format(col_val), fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)

plt.show()
```

```python
#percentile based outliers removal
def percentile_based_outlier(data, threshold=95):
    diff = (100 - threshold) / 2
    minval, maxval = np.percentile(data, [diff, 100 - diff])
    return (data < minval) | (data > maxval)

col_names = ['Year','Kilometers_Driven','Mileage(kmpl)','Engine(CC)','Seats','Price']
fig, ax = plt.subplots(len(col_names), figsize=(8,40))

for i, col_val in enumerate(col_names):
    x = car_data_hist[col_val][:1000]
    sb.distplot(x, ax=ax[i], rug=True, hist=False)
    outliers = x[percentile_based_outlier(x)]
    ax[i].plot(outliers, np.zeros_like(outliers), 'ro', clip_on=False)

    ax[i].set_title('Outlier detection - {}'.format(col_val), fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)

plt.show()
```
