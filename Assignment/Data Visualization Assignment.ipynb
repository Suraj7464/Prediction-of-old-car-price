## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
data=[['Delhi',82,48,15],['Mumbai',78,31,20],['Chennai',48,51,14],['Kolkata',46,29,13],['Hyderabad',39,18,9]]
data=pd.DataFrame(data,columns=['store','clothing','Equipment','Accesories'])
data.head()
```

```python
#attributes or columns 
columns=['Clothing','Equipment','Accessories']
# stores 
store1=np.array(data.store)
#clothing  
c = np.array(data.clothing)
# Equipment
e = np.array(data.Equipment)
#Accessories
a = np.array(data.Accesories)
index=[x for x,_ in enumerate(store1)]

# mark x and y label
plt.xlabel('Stores')
plt.ylabel('Range')

plt.bar(index, c, width=0.5, color='b', bottom = e+a)
plt.bar(index, e, width=0.5, color='r', bottom = a)
plt.bar(index, a, width=0.5, color='g')

# xticks with an angle of 90 degree
plt.xticks(index,store1,rotation='90')

# legend 
plt.legend(columns,title="Stores")

plt.title('Stacked Bar Graph')
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
data=sb.load_dataset("Titanic")
n=[data.pclass.unique()]
n.sort()
#data of non-survived passenger
data1=[data[(data['survived']==0)&(data['pclass']==1)].pclass.count(),data[(data['survived']==0)&(data['pclass']==2)].pclass.count(),data[(data['survived']==0)&(data['pclass']==3)].pclass.count()]
#data of survived passenger
data2=[data[(data['survived']==1)&(data['pclass']==1)].pclass.count(),data[(data['survived']==1)&(data['pclass']==2)].pclass.count(),data[(data['survived']==1)&(data['pclass']==3)].pclass.count()]
data.head()
```

```python

plt.title("Survival of passengers against the passanger class")
plt.xlabel("passenger class")
plt.ylabel("passenger Quantity")
n=3
index=np.arange(1,n+1)
bar_width=0.35
plt.bar(index,data1,bar_width,color='r',label='NO')
plt.bar(index+bar_width,data2,bar_width,color='g',label='YES')
plt.xticks(index,[1,2,3])
plt.legend(title='Survived',loc=2)
plt.show()


```

## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
country=['United States','Great Britain','India','Russia','Germany']
medal=[46,27,26,19,17]
ex=[0,0,0.2,0,0]
color=['c','yellowgreen','lightcoral','lightskyblue','gold']
# Plot
plt.pie(medal, explode=ex, labels=country, colors=color, shadow=True, pctdistance=0.6,startangle=150, autopct='%1.2f%%')
plt.axis('equal')
plt.title('Percentage distribution of medals among countries',color='b',size=15)
plt.show()
```

## Q.10 For Tips Dataset
- Load the dataset using seaborn
- Draw the Strip Plot between 'total bill' and 'day'
- Rotate xticks with an angle of 45 Degree

```python
# load the dataset
dataT=sb.load_dataset('tips')
dataT.head()
```

```python
# strip plot graph
x=sb.stripplot(x='day',y='total_bill',data=dataT,size=5)
plt.xticks(rotation=45)
x.set_xlabel('day',fontsize=15)
x.set_ylabel('total_bill',fontsize=15)
plt.show()
```
